### 4. Определите время года, в которое вероятность появления метеоритов максимальна. Постройте график.

```python
def func4(self):
    return str(self['Peak Brightness Date/Time (UT)'])[5:7]

df4 = df.copy()
df4['Peak Brightness Date/Time (UT)'] = df.apply(func4,axis=1)
df4 = df4.groupby('Peak Brightness Date/Time (UT)')[['Peak Brightness Date/Time (UT)']].count()
lst41 = ['Winter','Spring','Summer','Autumn']
lst42 = [(df4.iat[0,0]+df4.iat[1,0]+df4.iat[11,0])/len(df),(df4.iat[2,0]+df4.iat[3,0]+df4.iat[4,0])/len(df),(df4.iat[5,0]+df4.iat[6,0]+df4.iat[7,0])/len(df),(df4.iat[8,0]+df4.iat[9,0]+df4.iat[10,0])/len(df)]
print(str(lst41[lst42.index(max(lst42))]) + ": " + str(max(lst42)))
df42 = pd.DataFrame(dict([('Season',lst41), ('Value',lst42)])).plot(kind='bar',x='Season',y='Value',ylim=((min(lst42))*0.97,max(lst42)*1.03))
plt.show()
```

### 8. Определить месяц года, для которого появление метеоритов наиболее вероятно (построить необходимые графики).

```python
def func8(self):
    return str(self['Peak Brightness Date/Time (UT)'])[5:7]

df8 = df.copy()
df8['Peak Brightness Date/Time (UT)'] = df.apply(func8,axis=1)
df8 = df8.groupby('Peak Brightness Date/Time (UT)')[['Peak Brightness Date/Time (UT)']].count()
df8 = df8.sort_values(by='Peak Brightness Date/Time (UT)',ascending=False)
print('Minimum: month №' + str(df8.index[0]))
df8.plot(kind='bar',legend=None,ylim=(min(df8['Peak Brightness Date/Time (UT)']*0.9),max(df8['Peak Brightness Date/Time (UT)']*1.1)))
plt.show()
```
